```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
import os.path
import json
```

```python
df = pd.read_csv('cleaninsurance_testdata.csv')
df
```

```python
y = df.Response
x = df.drop(columns = ['Response'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
clf = MLPClassifier()

param_grid = {
    'solver': ['lbfgs', 'adam'], 
    'max_iter': list(range(1000,5000,500)), 
    'alpha': 10.0 ** -np.arange(1, 7)
}

CV_rfc = GridSearchCV(estimator=clf, param_grid=param_grid, cv= 5)
CV_rfc.fit(x_train.values, y_train.values)
print ('\n',CV_rfc.best_estimator_)
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(alpha=1e-06, max_iter=4500)
clf.fit(x_train.values, y_train.values)
```

```python
y_pred = clf.predict(x_test.values)
```

```python
from sklearn import metrics
print("Accuracy: ",metrics.accuracy_score(y_test, y_pred)*100)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
from mlxtend.plotting import plot_confusion_matrix
cm = confusion_matrix(y_test, y_pred)
fig, ax = plot_confusion_matrix(conf_mat=cm, figsize=(7, 7))
plt.title('Confusion Matrix', fontsize=16)
plt.show()
```

```python
import pickle
# create an iterator object with write permission - model.pkl
with open('neuralnet_pkl.sav', 'wb') as files:
    pickle.dump(clf, files)
```

```python
# load saved model
with open('neuralnet_pkl.sav' , 'rb') as f:
    lr = pickle.load(f)
```
