```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
import os.path
import json
```

```python
df = pd.read_csv('cleaninsurance_testdata.csv')
df
```

```python
first5k = df.iloc[0:5000]
save_response = first5k.Response
first5k.drop(columns=['Response'], inplace=True)
first5k.drop(columns=first5k.columns[0], axis=1, inplace=True)
print(first5k.info(verbose = True, show_counts=True))
```

```python
remaining = df.iloc[5001:59380]
remaining.drop(columns=remaining.columns[0], axis=1, inplace=True)
print(remaining.info(verbose=True, show_counts=True))
```

```python
y2 = remaining.Response
x2 = remaining.drop(columns = ['Response'])
print(type(x2),type(y2))
print(x2.info(verbose=True, show_counts=True))
```

```python
from sklearn.model_selection import train_test_split
x2_train, x2_test, y2_train, y2_test = train_test_split(x2, y2, test_size = 0.25, random_state = 0)
```

```python
# 0.0001 < gamma < 10
# 0.1 < c < 100
from sklearn.svm import SVC
svmlinear= SVC(kernel='rbf', C = 5, gamma = .01)
svmlinear.fit(x2_train, y2_train)
```

```python
y2_pred = svmlinear.predict(x2_test)
```

```python
from sklearn import metrics
print("Accuracy: ",metrics.accuracy_score(y2_test, y2_pred)*100)
```
